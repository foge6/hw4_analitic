## Задача 1.2. 
Изучите стоимости недвижимости


```python
import matplotlib.pyplot as plt
plt.hist(df['price'], bins = 100)
plt.title('Стоимость')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

## Условие 2: 2 задача
### Задача 2.1. 
Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
data = df['waterfront'].value_counts()
names = data.index
values = data.values

plt.figure(figsize=(3,3))
plt.bar(names,values)
plt.xticks();
```

### Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg');
```

На этом графике зависимость не прямая. Больше всего по цене квартиры, которые имеют, в среднем, 5 спален

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg');
```

На этом графике среднее распределение зависимости душевых от цены приближенно к прямой зависимости. Чем больше спален тем, в среднем, цена больше

```python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg');
```

На этом графике среднее распределение зависимости этажей от цены приближенно к прямой зависимости. Однако, больше ценятся квартиры на 2 этаже

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

На этом графике среднее распределение зависимости оценки от цены приближенно к прямой зависимости со сдвигом по оси y. 

```python
sns.jointplot(x=df['yr_built'], y=df['price'], kind='reg');
```

По графику видно, что цена дома никак не зависит от года постройки

```python
sns.jointplot(x=df['price'], y=df['condition'], kind='reg');
```

По графику видно, что больше ценятся квартиры с состоянием 3 и больше
